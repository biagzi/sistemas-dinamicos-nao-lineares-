### Figura 1

#### Simulação númerica

```python
logistic_map = lambda x, a: a * x * (1 - x)


def recursive_logistic_map(x, a, n):
    if n == 0:
        return x
    else:
        x = logistic_map(x, a)
        return recursive_logistic_map(x, a, n - 1)
```

```python
n = 1
a_list = [2.707, 3.414]
x_list = np.linspace(0, 1, 300)
a_container = {}
# Simulate
for a in a_list:
    next_x = recursive_logistic_map(x_list, a, n)
    a_container[a] = dict(zip(x_list, next_x))
    
# Plot
for a in a_list:
    x, y = zip(*a_container[a].items())
    plt.plot(x, y, 'x-', label="$a={}$".format(a))

plt.legend()
plt.show()
```
